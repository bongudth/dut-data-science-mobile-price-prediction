# Data Preprocessing

## ü™° Merge 4 raw_data files into 1

```python
import pandas as pd

raw_data_files = [pd.read_csv('../raw_data/cellphoneS.csv'),
                pd.read_csv('../raw_data/hoangHa1.csv'),
                pd.read_csv('../raw_data/hoangHa2.csv'),
                pd.read_csv('../raw_data/thegioididong.csv')]

data = pd.concat(raw_data_files, ignore_index=True)

data.head()
```

```python
data.drop(['Unnamed: 0', 'Name'], axis=1, inplace=True)

data.head()
```

```python
import re

# parse float from string
def parse_float(string):
    string = str(string)
    string = string.strip()
    if string == '':
        return None
    
    string = string.replace(',', '.')
    string = string.replace('"', '')
    size = re.findall('\d+\.\d+', string)
    if len(size) > 0:
        return float(size[0])
    return ''

dataScreenSize = data['Screen size']
dataScreenSize = dataScreenSize.apply(parse_float)

# get data randomly
dataScreenSize = dataScreenSize.sample(frac=1)
dataScreenSize.head()
```

```python
dataScreenType = data['Screen type']
# list of screen type

def filter_list_screen_type(string):
    string = str(string)
    screen_types = ['Super AMOLED', 'IPS LCD', 'OLED', 'AMOLED', 'LCD', 'TFT LCD', 'PLS TFT LCD', 'QQVGA', 'K√≠nh m√†n h√¨nh: Sapphire', 'Liquid Retina']
    if string in screen_types:
        return string

    for screen_type in screen_types:
        if screen_type in string:
            return screen_type

    return ''

dataScreenType = dataScreenType.apply(filter_list_screen_type)
dataScreenType.head(20)
dataScreenType.value_counts()
```

```python
dataROM = data['ROM']
dataROM.head(50)
def filter_rom(string):
    string = str(string)
    string = string.replace(' ', '')
    regex = re.compile('\d+[GB|MB|TB]')
    size = regex.findall(string)
    if len(size) > 0:
        return size[0]
    return ''

dataROM = dataROM.apply(filter_rom)
dataROM.value_counts()
```

```python
dataRAM = data['RAM']
dataRAM.head(50)
def filter_rom(string):
    string = str(string)
    string = string.replace(' ', '')
    regex = re.compile('\d+[GB|MB|TB]')
    size = regex.findall(string)
    if len(size) > 0:
        return size[0]
    return ''

dataRAM = dataRAM.apply(filter_rom)
dataRAM.value_counts()
```

```python
dataBattery = data['Batery']
dataBattery.head(50)
def filter_rom(string):
    string = str(string)
    string = string.replace(' ', '')
    regex = re.compile('\d+mAh')
    matched_str = regex.findall(string)
    if len(matched_str) > 0:
        return matched_str[0].replace('mAh', '')
    
    return ''

dataBattery = dataBattery.apply(filter_rom)
dataBattery.value_counts()
```

```python
data_refresh_rate = data['Refesh rate']
data_refresh_rate.head(50)
def filter_refresh_rate(string):
    string = str(string)
    string = string.replace(' ', '')
    regex = re.compile('\d+Hz')
    matched_str = regex.findall(string)
    if len(matched_str) > 0:
        return matched_str[0]
    return string

data_refresh_rate = data_refresh_rate.apply(filter_refresh_rate)
data_refresh_rate.value_counts()
```
